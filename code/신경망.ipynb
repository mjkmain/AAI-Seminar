```python
#1. 학습률, epoch을 변경하면서 정답이 나오도록 조절 
import numpy as np
InputLayerSize, hiddenLayerSize, outputLayerSize = 2, 3, 1

X = np.array([[0, 0], [0, 1], [1, 0], [1, 1]])
Y = np.array([[0],[1],[1],[0]])

def sigmoid(x):
    return 1/(1+np.exp(-x))
def sigmoid_diff(x):
    return x*(1-x)


#hidden_Weights = np.random.randn(InputLayerSize, hiddenLayerSize)
hidden_Weights = np.random.uniform(size=(InputLayerSize,hiddenLayerSize))
print("----------------Hidden Layer Weights----------------")
print(hidden_Weights)
print(hidden_Weights.shape)
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
#Output_Weights = np.random.randn(hiddenLayerSize, outputLayerSize)
Output_Weights = np.random.uniform(size=(hiddenLayerSize,outputLayerSize))
print(Output_Weights)
print(Output_Weights.shape)
print("xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx\n")

epochs = int(input("epoch값 입력 : ")) #epoch
L = float(input("Learning rate값 입력 :")) #LR

for i in range(epochs):
    Hidden_result = sigmoid(np.dot(X, hidden_Weights)) #4by2*2by3 -> 4by3
    Output = np.dot(Hidden_result, Output_Weights)     #4by3*3by1 -> 4by1
    Error = Y - Output
    dZ = Error*L
    Output_Weights += Hidden_result.T.dot(dZ)          #3by4*4by1 -> 3by1
    dHidden_result = dZ.dot(Output_Weights.T)*sigmoid_diff(Hidden_result)
    hidden_Weights += X.T.dot(dHidden_result)
m = np.sum(Output)/len(X)
for i in range(len(X)):
    if Output[i][0] < m:
        Output[i][0] = 0
    else:
        Output[i][0] = 1
print("\n XOR 2 inputs")
print(Output)
```
